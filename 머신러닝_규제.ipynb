## Ridge, Lasso 시각화

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,4))

rng = np.linspace(-10,10,100)
mse = (0.5*(rng-3))**2 + 30
l2 = rng**2
l1 = 5*np.abs(rng)
ridge = mse + l2
lasso = mse + l1

plt.subplot(1,2,1)
plt.plot(rng,mse,label='MSE')
plt.plot(rng,l2,'--',label='L2')
plt.plot(rng,ridge, lw=2, label='Ridge (=MSE+L2)')
plt.xlabel('w'); plt.ylabel('Error')
plt.title('Ridge(=MSE+L2)')
plt.legend()

plt.subplot(1,2,2)
plt.plot(rng,mse,label='MSE')
plt.plot(rng,l1,'--',label='L1')
plt.plot(rng,lasso, lw=2, label='Lasso (=MSE+L1)')
plt.xlabel('w'); plt.ylabel('Error')
plt.title('Lasso (=MSE+L1)')
plt.legend()

plt.show()
```

## Ridge Regression
* 선형 모델(MSE 최소화) + 전체적으로 가중치 최소화
* alpha 값으로 회귀계수를 압박 (alpha=0, 선형회귀와 동일)

```python
# Ridge Regression
from sklearn.linear_model import Ridge
model = Ridge(alpha=0.1)
x = [[-1,-1],[0,0],[1,1]]
y = [-1,0,1]
model.fit(x,y)

model.coef_ , model.intercept_
```

```python
model.score(x, y)
```

```python
model.predict([[3,-3]])
```

## ElasticNet Regression

* 릿지와 라쏘 페널티 결합
* Lasso는 ElasticNet(l1_ratio=1.0)와 동일

```python
# ElasticNet Regression

from sklearn.linear_model import ElasticNet
model = ElasticNet(alpha=0.1)
x = [[-1,-1],[0,0],[1,1]]
y = [-1,0,1]

model.fit(x,y)

model.coef_ , model.intercept_
```

```python
pred_x = [[3,-3]]
model.predict(pred_x)
```
