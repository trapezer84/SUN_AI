> PCA 모듈의 pca.explained_variance_ratio_ 속성에 대해 설명하고 예시를 작성해줘



`PCA` (Principal Component Analysis, 주성분 분석)는 데이터셋을 고차원에서 저차원으로 축소하는 기법 중 하나입니다. `PCA`는 원래 차원의 데이터를 선형결합을 통해 새로운 축으로 변경한 후 최대한 분산이 보존되도록 구성합니다. 이때 새롭게 변환된 축들은 기존 데이터의 주요 구조를 나타내는데, 이러한 축을 주성분이라고 합니다.

`PCA`를 수행하면 주어진 데이터셋의 주성분의 개수, 그리고 각 주성분이 데이터셋의 분산을 얼마나 잘 설명하는지를 알 수 있습니다. 이 때 `explained_variance_ratio_` 속성이 사용됩니다. `explained_variance_ratio_`는 주성분의 분산에서 전체 분산에서의 대비 비율을 나타냅니다.

예를 들어, 아래와 같이 `sklearn`의 `PCA`를 이용하여 데이터셋을 2차원으로 축소해보겠습니다.

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA

# 임의의 데이터셋 생성
X, y = make_blobs(n_samples=200, n_features=5, centers=3, random_state=42)
# PCA를 이용해 데이터셋을 2차원으로 축소
pca = PCA(n_components=2)
pca.fit(X)
X_pca = pca.transform(X)

# 주성분의 분산에서 전체 분산에서의 대비 비율 출력
print("주성분의 분산에서 전체 분산에서의 대비 비율: {}".format(pca.explained_variance_ratio_))
```

위 코드를 실행하면 아래와 같은 출력을 볼 수 있습니다.

```
주성분의 분산에서 전체 분산에서의 대비 비율: [0.56629431 0.29539815]
```

해석해보면 첫 번째 주성분은 데이터셋의 분산 구조 중 약 56.6%를 설명하고, 두 번째 주성분은 29.5%를 설명한다고 해석할 수 있습니다. 즉, 이 두 주성분을 사용하여 2차원으로 축소했을 때, 전체 데이터셋의 약 86.1%를 설명할 수 있는 것입니다.

````python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA

# 임의의 데이터셋 생성
X, y = make_blobs(n_samples=200, n_features=5, centers=3, random_state=42)
# PCA를 이용해 데이터셋을 2차원으로 축소
pca = PCA(n_components=2)
pca.fit(X)
X_pca = pca.transform(X)

# 주성분의 분산에서 전체 분산에서의 대비 비율 출력
print("주성분의 분산에서 전체 분산에서의 대비 비율: {}".format(pca.explained_variance_ratio_))
````

````python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 배열 시각화를 위해 seaborn 패키지를 불러옵니다.
import seaborn as sns

# 임의의 데이터셋 생성
X, y = make_blobs(n_samples=200, n_features=5, centers=3, random_state=42)

# PCA를 이용해 데이터셋을 2차원으로 축소
pca = PCA(n_components=2)
pca.fit(X)
X_pca = pca.transform(X)

# 주성분의 분산에서 전체 분산에서의 대비 비율 출력
print("주성분의 분산에서 전체 분산에서의 대비 비율: {}".format(pca.explained_variance_ratio_))

# 2차원으로 축소한 데이터셋 시각화
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='Set2')
plt.show()
````

> mnist 데이터를 사용하고, PCA 알고리즘으로 시각화 하는 예제 작성해줘  


````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA

# MNIST 데이터셋을 불러옴
mnist = fetch_openml('mnist_784', version=1, data_home="./data/")
X, y = mnist['data'], mnist['target']
y = y.astype(int)

# PCA를 이용해 2차원으로 차원 축소
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 2차원으로 축소한 데이터셋 시각화
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, legend='full', palette='Set2')
plt.show()
````

> iris 데이터를 사용하고, TSNE와  PCA 알고리즘으로 시각화해서 비교하는 예제 작성해줘

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# iris 데이터셋을 불러옴
iris = load_iris()
X, y = iris.data, iris.target

# PCA를 이용해 2차원으로 차원 축소
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# TSNE를 이용해 2차원으로 차원 축소
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)

# PCA와 TSNE 시각화
fig, axs = plt.subplots(ncols=2, figsize=(12, 4))

sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, legend='full', palette='Set2', ax=axs[0])
axs[0].set_title('PCA')

sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=y, legend='full', palette='Set2', ax=axs[1])
axs[1].set_title('TSNE')

plt.show()
````
