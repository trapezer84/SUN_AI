## 단순 선형 회귀 예제

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 데이터 생성
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = 3*X + 2 + np.random.randn(80, 1)

# 단순 선형 회귀 모델 훈련
model = LinearRegression()
model.fit(X, y)

# 예측 결과 시각화
plt.scatter(X, y, color='blue', label='Data with Noise')
plt.plot(X, 3*X + 2, color='red', label='True Relationship')
plt.plot(X, model.predict(X), color='green', label='Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

## 다중 선형 회귀 예제:

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 가상 데이터 생성
np.random.seed(0)
X = np.random.rand(100, 2)  # 각 열은 크기와 방의 개수
y = 5*X[:, 0] + 3*X[:, 1] + 2 + np.random.randn(100)

# 다중 선형 회귀 모델 훈련
model = LinearRegression()
model.fit(X, y)

# 예측 예시
new_data = np.array([[0.6, 3]])  # 크기: 0.6, 방의 개수: 3
predicted_price = model.predict(new_data)
print("Predicted Price:", predicted_price)
```
