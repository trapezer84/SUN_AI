# 머신러닝 - 다항회귀(PolynomialFeatures)

**2019-2023 [FinanceData.KR]()**


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 데이터 생성
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = 0.5*X**2 + X + 2 + np.random.randn(80, 1)

# 다항 선형 회귀 모델 훈련
degree = 2  # 다항 차수
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)
model = LinearRegression()
model.fit(X_poly, y)

# 예측 결과 시각화
plt.scatter(X, y, color='blue', label='Data with Noise')
plt.plot(X, 0.5*X**2 + X + 2, color='red', label='True Relationship')
plt.plot(X, model.predict(X_poly), color='green', label=f'Polynomial Regression (Degree {degree})')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Linear Regression')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# 가상 데이터 생성
np.random.seed(0)
X = np.sort(10 * np.random.rand(80, 1), axis=0)
y = np.sin(X) + np.random.randn(80, 1)

degree = 10  # 다항 차수
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)
model = LinearRegression()
model.fit(X_poly, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 시각화를 위한 그리드 생성
X_range = np.linspace(min(X), max(X), 100).reshape(-1, 1)
X_poly_range = poly_features.transform(X_range)
y_pred_range = model.predict(X_poly_range)

# 데이터와 회귀 곡선 시각화
plt.scatter(X, y, color='blue', label='Data with Noise')
plt.plot(X_range, np.sin(X_range), color='red', label='True Relationship')
plt.plot(X_range, y_pred_range, color='green', label=f'Polynomial Regression (Degree {degree})')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression on Complex Data')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# 가상 데이터 생성
np.random.seed(0)
X = np.sort(10 * np.random.rand(80, 1), axis=0)
y = np.sin(X) + np.random.randn(80, 1)

# 시각화
plt.figure()
plt.scatter(X, y, color='blue', label='Data with Noise')
plt.plot(X, np.sin(X), color='red', label='True Relationship')

# 다양한 degree 값에 대한 다항 회귀 시각화
for degree in [1, 3, 15]:
    poly_features = PolynomialFeatures(degree=degree)
    X_poly = poly_features.fit_transform(X)
    model = LinearRegression()
    model.fit(X_poly, y)
    y_pred = model.predict(X_poly)
    plt.plot(X, y_pred, label=f'Degree {degree}')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Effect of Degree in Polynomial Regression')
plt.legend()
plt.show()
```
